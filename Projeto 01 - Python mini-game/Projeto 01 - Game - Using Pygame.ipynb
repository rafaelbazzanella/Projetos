```python
import pygame
import time
import random
import time


class Character:
    '''
    Classe que serve para todos os personagens e objetos que interagem entre si no jogo 
    '''
    def __init__(self,tipo:str, pos_x:int, pos_y:int, width:int, height:int, speed:int=5, color:tuple=(100,100,100),imagem:str=''):
        '''
        Aloca em variáveis dentro da classe cada característica de um personagem ou item
        '''
        self.tipo=tipo
        self.pos_x =pos_x
        self.pos_y =pos_y
        self.width=width
        self.height=height
        self.speed=speed
        self.color=color
        self.imagem=imagem
        self.gameDisplay = pygame.display.set_mode((display_width,display_height))

    def hit_box_rect(a):
        ''' Essa função cria a área de hitbox(área aonde os personagens podem colidir) de um Character, 
            caso não existe uma imagem definida, será criado um retângulo cinza para sinalizar o movimento
            do Character criado.        
        '''
        if a.imagem =='':
            pygame.draw.rect(a.gameDisplay, a.color, [a.pos_x, a.pos_y, a.width, a.height])
        elif a.imagem !='':
            foto = pygame.image.load(a.imagem)
            gameDisplay.blit(foto, (a.pos_x,a.pos_y))
       

  
    def reborn_y(who,sentido:int=0):
        '''Essa função verifica se o personagem passou do canto inferior da tela e faz 
        reaparecer no canto superior da tela caso o sentido escolhido seja 0 e caso
        o sentido escolhido seja 1, fará o item resnacer no canto inferior.'''
        if sentido==0:
            if who.pos_y > display_height:
                who.pos_y = 0 - who.height
                who.pos_x = random.randrange(0,display_width)
        elif sentido ==1:
            if who.pos_y < -300:
                who.pos_y = display_height + who.height
                who.pos_x = random.randrange(0,display_width)
            
    def reborn_x(who):
        '''Essa função verifica se o personagem passou do canto lateral esquerdo da tela e faz ele 
        reaparecer no canto lateral direito da tela.'''
        if who.pos_x < 0:
            who.pos_x = display_width + who.height
            who.pos_y = random.randrange(0,display_height)
    def reborn_y_vida(who):
        '''Essa função faz um personagem resnacer automaticatimente no canto superior da tela.'''
        who.pos_y = -1000 - who.height
        who.pos_x = random.randrange(0,display_width)

def collision(who1,who2):
    '''
    Função que compara a posição de um personagem em relação ao outro e retorna true caso exista colisão entre eles
    '''
    if (who1.pos_x+who1.width) > (who2.pos_x) and (who1.pos_x) < (who2.pos_x+who2.width):
        if (who1.pos_y+who1.height) > (who2.pos_y) and (who1.pos_y) < (who2.pos_y+who2.height): 
            return True

pygame.init()

display_width = 800
display_height = 600
gameDisplay = pygame.display.set_mode((display_width,display_height)) #tamanho da tela 
pygame.display.set_caption('My First Awesome Game - Avoid the dangers of life! ') #nome da aba do jogo
clock = pygame.time.Clock() #Para definir qual velocidade de frames per second o jogo terá



colors={'black':(0,0,0),'white':(255,255,255),'red':(255,0,0),'green':(0,255,0),
        'yellow':(255,255,0),'navy':(0,0,128),'deepskyblue':(0,191,255),'azulvioleta':(159,95,159)

       }



def text_display(display_size:tuple,text:str,position:tuple=(0,0),color:tuple =(255,255,255)):
    ''' This function display a text somewhere you choose with the color you want!
    '''
    screen_display=pygame.display.set_mode(display_size)
    font = pygame.font.SysFont(None, 50)
    text = font.render(str(text), True, color)
    screen_display.blit(text,position)

def player_name(position:tuple=(0,0),color:tuple =(255,255,255)):
    '''This function gets the user name and return a string with the choosen name as string'''
    string_name=str()
    check= True
    
    for event in pygame.event.get():
    
        while check: 
            pressed = pygame.key.get_pressed()
            keys = pygame.key.get_pressed()
            for event in pygame.event.get(): #getting the key pressed           
                if event.type == pygame.KEYDOWN:
                    print(event.key)
                    key = pygame.key.name(event.key)  # id da tecla
                    if len(key) == 1:  # pegando qualquer tecla do teclado                    
                        string_name += key
                    elif key == "backspace": #deletando última letr
                        string_name = string_name[:len(string_name) - 1]
                    elif event.key == 13:  # Enter para acabar.
                        string_name.capitalize()
                        check = False     
            text_display((display_width,display_height),'Player: '+string_name,(display_width/3,display_height/2))
            pygame.display.flip() 

        
         
            
            if event.type == pygame.QUIT: #if que verifica se fecharam o jogo e finaliza o python
                pygame.quit() 
                quit()
        
        
    
    
    return str(string_name)        
 

       
def score_display(pontos,where:str =(0,0)):
    '''
    Essa função mostra o score, level e nome do player no canto superior esquerdo da tela.
    '''
    font = pygame.font.SysFont(None, 25)
    global score
    global level
    
    
    text = font.render("Score: "+str(int(pontos))+" Level: "+str(level)+" Player: "+str(name_player), True, colors['black'])
    gameDisplay.blit(text,where)

    
  
    

    
def text_objects(text, font):
    '''
    Renderiza uma mensagem porém acaba com o jogo
    '''
    textSurface = font.render(text, True, colors['black'])
    return textSurface, textSurface.get_rect()


def message_display(text):
    '''
    Função que permite enviar uma mensagem ao jogador
    '''
    largeText = pygame.font.Font('freesansbold.ttf',50) #fonte e tamanho
    TextSurf, TextRect = text_objects(text, largeText) #definir a superfície e colocar num retângulo
    TextRect.center = ((display_width/2),(display_height/4)) #posicionamento da mensagem
    gameDisplay.blit(TextSurf, TextRect) #mostra para o jogador
    pygame.display.update()
    time.sleep(4)    #tempo da mensagem
    game_loop()      #chama a função para começar outro jogo

    
def you_lost():
    '''
    Mensagem para o jogar dizendo que perdeu
    '''
    score_display(score)
    message_display('You lost! Try again!') 
    
    
    
start = time.time() # para fazer o score através do tempo vivo mais extra durante o jogo




def game_loop(): #criado um loop do jogo para que só saia dele quando o jogador perder
    global extra_points
    global score
    global start
    global name_player
    global level
    global vida
    
    game_over = False
    score=0 #pontuação inicial zerada
    extra_points =0 #pontuação extra inicial
    
    x_change = 0 #variável para mudança de posição
    y_change = 0
    vida=3
    
    name_player=str(player_name()) #Atribuindo a uma variável para apenas roda 1 vez
 
    #contadores de desvios
    thing_count = 1
    dodged = 0
    
    ###### inserindo os personagens
    enemy_1=Character('Monster',random.randrange(0, display_width-138),-600,138,168,3,imagem='enemy_1.png') #########criando um personamge aqui
    enemy_2=Character('Monster',random.randrange(0, display_width-83),display_height+60,83,131,5,imagem='enemy_2.png')
    enemy_3=Character('Monster',random.randrange(0, display_width-150),-600,150,81,7,imagem='enemy_3.png')
    personagem_1 =Character('Personagem',(display_width * 0.45),(display_height * 0.7),90,140,imagem='carteira_reduzida.png')
    item_bonus_1 = Character('Item',random.randrange(0, display_width-50),-1000,50,50,7,imagem='item_bonus_1.png')
    item_bonus_2 = Character('Item',display_width+1000,random.randrange(0, display_height-80),50,50,10,imagem='item_bonus_2.png')
    vida_1 = Character('Vida',display_width-130,0,0,0,0,imagem='vida_1.png')
    vida_2 = Character('Vida',display_width-130,0,0,0,0,imagem='vida_2.png')
    vida_3 = Character('Vida',display_width-130,0,0,0,0,imagem='vida_3.png')
    
    while not game_over: #rodar este código enquanto o jogo não acaba
        global level
           
        
        
        if score <=300:
            level = 1
        elif score <=800:
            level = 2
        else:
            level = 3
            
        
        
        for event in pygame.event.get():
            #print(event) #se quiser entender tudo que está acontecendo
            if event.type == pygame.QUIT: #if que verifica se fecharam o jogo e finaliza o python
                pygame.quit() 
                quit()
                
            if event.type == pygame.KEYDOWN: 
                if event.key == pygame.K_LEFT:
                    x_change = -5
                elif event.key == pygame.K_RIGHT:
                    x_change = 5
                elif event.key == pygame.K_UP:
                    y_change = -5
                elif event.key == pygame.K_DOWN:
                    y_change = 5
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                    x_change = 0
                if event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                    y_change = 0
                    
         #altera posição do personagem no eixo x e não permite passar das bordas laterais    
        if (personagem_1.pos_x < display_width - personagem_1.width) and personagem_1.pos_x > 0: 
            personagem_1.pos_x += x_change
        elif (personagem_1.pos_x <= display_width - personagem_1.width) and x_change < 0 and personagem_1.pos_x > 0:
            personagem_1.pos_x += x_change
        elif personagem_1.pos_x <= 0 and x_change > 0:
            personagem_1.pos_x +=x_change
        
        #altera posição do personagem no eixo y e não permite passar da borda inferior e superior
        if (personagem_1.pos_y < display_height - personagem_1.height) and personagem_1.pos_y > 0: #altera posição do personagem no eixo x e não deixa passar da borda
            personagem_1.pos_y += y_change
        elif (personagem_1.pos_y <= display_height - personagem_1.height) and y_change < 0 and personagem_1.pos_y > 0:
            personagem_1.pos_y += y_change
        elif personagem_1.pos_y <= 0 and y_change > 0:
            personagem_1.pos_y +=y_change    
                
         
        #Comando para trocar a cor do fundo para cada nivel 
        if level == 1:
            gameDisplay.fill(colors['deepskyblue']) #cor do fundo da tela                       
        elif level  ==2:
            gameDisplay.fill(colors['green'])
        else:
            gameDisplay.fill(colors['azulvioleta'])
       
        
        if collision(personagem_1,item_bonus_1):
            extra_points +=2
            if level ==2:
                    gameDisplay.fill(colors['yellow'])
            else:
                gameDisplay.fill(colors['green'])
        elif collision(personagem_1,item_bonus_2):
            extra_points +=4
            if level ==2:
                gameDisplay.fill(colors['yellow'])
            else:
                gameDisplay.fill(colors['green'])
        elif collision(personagem_1,enemy_1):
            extra_points -=1
            if int(score)%2==0:
                gameDisplay.fill(colors['red']) 
            elif int(score)%3==0:
                gameDisplay.fill(colors['black'])
        elif collision(personagem_1,enemy_2):
            extra_points -=1
            if int(score)%2==0:
                gameDisplay.fill(colors['red']) 
            elif int(score)%3==0:
                gameDisplay.fill(colors['black'])
        elif collision(personagem_1,enemy_3): #comando para quando inimigo 3 encostar, dizer que perdeu o jogo
            enemy_3.reborn_y_vida()
            vida-=1            
            
            if vida ==0:
                start = time.time() + 4
                you_lost()
       
    
        if vida ==3:
            vida_3.hit_box_rect()
        elif vida==2:
            vida_2.hit_box_rect()
        elif vida==1:
            vida_1.hit_box_rect()
        
        
        
        #comando para nascer
        enemy_1.hit_box_rect()
        enemy_1.pos_y +=enemy_1.speed #comando para dar movimento
        enemy_1.reborn_y()
        
        if level >=2:
            enemy_2.hit_box_rect()
            enemy_2.pos_y -=enemy_2.speed #comando para dar movimento
            enemy_2.reborn_y(sentido=1)
            
            item_bonus_2.hit_box_rect()
            item_bonus_2.pos_x -=item_bonus_1.speed #comando para dar movimento
            item_bonus_2.reborn_x()
            
        if level >=3:
            enemy_3.hit_box_rect()
            enemy_3.pos_y +=enemy_3.speed #comando para dar movimento
            enemy_3.reborn_y()
        
        
        item_bonus_1.hit_box_rect()
        item_bonus_1.pos_y +=item_bonus_1.speed #comando para dar movimento
        item_bonus_1.reborn_y()
        
        personagem_1.hit_box_rect()
        
        
        
        
                
        
        end = time.time() #acaba de contar o relógio
        score = extra_points + (end - start) # scoretotal        
        score_display(score)   #mostrando o placar
        
        pygame.display.update() #atualiza a tela, se não passar parâmetro, atualiza a tela toda
        clock.tick(60)

        
        
        
game_loop()    
pygame.quit() 
quit()
```
